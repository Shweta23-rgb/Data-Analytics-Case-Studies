### Data Exploration: 

In my opinion, there are two stages in Data Exploration. One, univariate and two bivariate analysis. 

Now, Whats the difference between the two: 

1. **Univariate Analysis:** Looking at one variable at a time. 
2. **Bivariate Analysis:** Looking at two things together, and seeing how one variable affects another. This is to find relationships between two variables. 

If I have to explain like I would to a 5 year old: 

**Univariate Analysis: Do you have more candy?**

**Bivariate Analysis: Does older kids get more candy?**



```python
import matplotlib.pyplot as plt
import seaborn as sns

num_vars = ['age', 'viral load', 'severity level', 'hospitalization charges']

for col in num_vars:
    print(f"\nSummary of {col}:\n", df[col].describe())
    
    plt.figure(figsize=(14, 5))
    
    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    
    plt.tight_layout()
    plt.show()
```

### Interpretaions:

**Age:** 

There's a big spike around age 20, meaning more patients are aged 20 than any other age group.

Rest of the ages are fairly evenly distributed, with some small fluctuations.

The curve is slightly left-skewed due to that spike.

No outliers. 

**Viral Load** 

The data forms a bell-shaped curve → approximately normally distributed (nice!)

The mean is approximately equal to median visually, and the distribution is centered around ~10.

The middle 50% of the data (IQR) lies between ~8 and ~12.

Median viral load is just under 10.

Yes, there are outliers, the dots we see to the far right (values > 15) are mild outliers. [if I were to model it, I would take a look at them before modelling]


**Severity Level**

The histogram is clearly right skewed. 

Most patients are in a severity level of 0-2. Very few patients have a severity level of 4-5 (rare!)

There are no major/ extreme outliers here. Median appears around 1, meaning half the patients had severity ≤1




```python
### Univariate Analysis: For Categorical Columns: 

### We count frequency using value_counts() and plot bar graphs to check the frequency.

cat_vars = ['sex', 'smoker', 'region']

for col in cat_vars:
    print(f"\nValue counts of {col}:\n", df[col].value_counts())

    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f'Count of {col}')
    plt.show()
```

**Interpretations** 

1. Female approx equal to male (great!, No class imbalence here.)
2. Non-smoker are higher,they dominate the dataset(~75% of them are non smokers) [Could still be very meaningful, especially if smoking is linked to higher viral load or charges]
3. All 4 regions are evenly represneted
                                    

### Bivariate Analysis: 

```python
sns.scatterplot(x='viral load', y='hospitalization charges', data=df)
print("Correlation:", df['viral load'].corr(df['hospitalization charges']))
```

**My Interpretation** 

There's a positive trend: As viral load increases, hospitalization charges tend to increase. But this is not perfectly linear - theres a lot spread. 

There is a dense cluster between 8-12 viral load with hospitalisation charges around 40k, possibly standard core charges. 

To be sure lets check the correlation value

```python
correl_viral_hp = df['viral load'].corr(df['hospitalization charges'])
print(correl_viral_hp)
```

0.2 meaning it is a weak positive correlation.

There’s some relationship, but it’s not strong or linear.

**Meaning:** higher viral load may lead to higher charges, but many other factors are involved.

### Bivariate Analysis: Categorical Columns: 



```python
# Boxplot
sns.boxplot(x='smoker', y='hospitalization charges', data=df)
plt.title("Smoker vs Hospitalization Charges")
plt.show()

# Group means
print(df.groupby('smoker')['hospitalization charges'].mean())
```

```python
# Boxplot
sns.boxplot(x='sex', y='hospitalization charges', data=df)
plt.title("Sex vs Hospitalization Charges")
plt.show()

# Group means
print(df.groupby('sex')['hospitalization charges'].mean())
```

```python
# Boxplot
sns.boxplot(x='region', y='hospitalization charges', data=df)
plt.title("Region vs Hospitalization Charges")
plt.show()

# Group means
print(df.groupby('region')['hospitalization charges'].mean())
```

**My Interpretations** 

Smokers clearly show a higher median, greater spread, and more higher-end outliers. Smoking is likely linked to higher complications → more expensive treatment.

Sex is not a strong driver of hospitalization cost.

Region does not significantly affect hospitalization charges.
