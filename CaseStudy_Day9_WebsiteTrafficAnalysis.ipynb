### Clickrate by Links: 

```python
# Clicks and pageviews per link
clicks_by_link = click_df.groupby('linkid').size()
pageviews_by_link = df_pageview.groupby('linkid').size()

# Join them together
link_stats = pd.DataFrame({
    'clicks': clicks_by_link,
    'pageviews': pageviews_by_link
}).fillna(0)

link_stats['click_rate'] = link_stats['clicks'] / link_stats['pageviews']


print(link_stats)
```

### Is there any correlation between clicks and previews on a link? Is it significant? How large is the effect? Make sure to at least test for potential linear as well as categorical (think binary) relationships between both variables.

```python
df['event'].value_counts()
```

```python
#Previews By Link: 

previews_by_link = df_previews.groupby('linkid').size()


#Add to stats
link_stats['previews'] = previews_by_link
link_stats = link_stats.fillna(0)
```

```python
#Linear Correlation: 

from scipy.stats import pearsonr

corr, pval = pearsonr(link_stats['clicks'], link_stats['previews'])
print(f"Pearson Correlation: {corr:.3f}, P-value: {pval:.5f}")
```

```python
#Binary Preview Impact: 

link_stats['has_preview'] = (link_stats['previews'] > 0).astype(int)

# Clicks by binary preview status
link_stats.groupby('has_preview')['clicks'].describe()
```

```python
from scipy.stats import ttest_ind

group1 = link_stats[link_stats['has_preview'] == 1]['clicks']
group2 = link_stats[link_stats['has_preview'] == 0]['clicks']

t_stat, p_val = ttest_ind(group1, group2, equal_var=False)
print(f"T-stat: {t_stat:.2f}, P-value: {p_val:.5f}")
```

Our analysis revealed a Pearson correlation coefficient of 0.98 between the number of previews and the number of clicks on a link, with a p-value < 0.00001. This implies an extremely strong and statistically significant positive linear relationship. Therefore, previews are a key engagement step and likely a good indicator of user intent to click.



Final Insight: 
    
Links with previews tend to get more clicks on average (over 70x higher), and this result complements the earlier Pearson correlation (0.98). While not every previewed link performs well, the overall data clearly supports that having a preview dramatically increases the chance of user interaction.
