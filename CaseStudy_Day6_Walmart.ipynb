```python
#Number of transactions made by gender: 

import matplotlib.pyplot as plt
import seaborn as sns

gender_counts = df['Gender'].value_counts()
print(gender_counts)

sns.countplot(data=df, x='Gender')
plt.title("Number of Transactions by Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Total Purchase amount by gender: 

total_purchase_by_gender = df.groupby('Gender')['Purchase'].sum().sort_values(ascending=False)
print(total_purchase_by_gender)

sns.barplot(x=total_purchase_by_gender.index, y=total_purchase_by_gender.values)
plt.title("Total Purchase Amount by Gender")
plt.ylabel("Total Purchase ($)")
plt.show()
```

```python
#Avergae Purchase Amount: 

purchase_stats = df.groupby('Gender')['Purchase'].agg(['mean', 'median']).round(2)
print(purchase_stats)

purchase_stats.plot(kind='bar')
plt.title("Average and Median Purchase Amount by Gender")
plt.ylabel("Purchase Amount ($)")
plt.xticks(rotation=0)
plt.show()
```

```python
#Distribution of Purchase Amounts by Gender: 

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Gender', y='Purchase')
plt.title("Purchase Amount Distribution by Gender (Boxplot)")
plt.show()

plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='Purchase', hue='Gender', fill=True, alpha=0.4)
plt.title("Distribution of Purchase Amounts by Gender (KDE Plot)")
plt.show()
```

#### Transaction Volume
Male customers made ~3x more transactions than female customers.

This indicates greater shopping frequency or engagement among male customers during this period.

#### Total Spending
The total purchase amount by male customers is more than 3x that of female customers.

This could be due to a higher number of male shoppers and possibly higher-ticket purchases.

#### Average & Median Spend
Metric	Male	Female
Mean	Higher	Lower
Median	Higher	Lower

On average, men not only shop more often, but also spend more per transaction.

The gap in both mean and median confirms it's not just outliers—spending is genuinely higher across the board.



#### BY AGE: 

```python
# Make sure 'Age' is treated as a categorical with correct order
from pandas.api.types import CategoricalDtype

age_order = CategoricalDtype(
    categories=['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+'],
    ordered=True
)
df['Age'] = df['Age'].astype(age_order)
```

```python
#Total Purchase BY Age Group: 

age_group_purchase = df.groupby('Age')['Purchase'].sum().sort_values(ascending=False)
print(age_group_purchase)

# Plot
sns.barplot(x=age_group_purchase.index, y=age_group_purchase.values)
plt.title("Total Purchase Amount by Age Group")
plt.ylabel("Total Purchase ($)")
plt.xlabel("Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
age_gender_purchase = df.groupby(['Age', 'Gender'])['Purchase'].sum().reset_index()

# Plot as grouped bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=age_gender_purchase, x='Age', y='Purchase', hue='Gender')
plt.title("Total Purchase by Age Group and Gender")
plt.ylabel("Total Purchase ($)")
plt.xlabel("Age Group")
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

#### Spending Patterns by Age Group
1️⃣ Total Purchase by Age Group
The 26–35 age group is the dominant spender, contributing the highest total purchase volume by a large margin.

This group is likely in their peak earning/spending years, making them a prime target for promotions.

2️⃣ Gender Comparison within Age Groups
In every age category, male customers outspent female customers — and significantly so.

The gender gap is most noticeable in the 26–35 and 36–45 age brackets.

This suggests that male consumers not only shop more frequently (as seen earlier), but also drive the highest spending across all age bands.

```python
### Product Category: 

# Group by product category and gender
category_gender = df.groupby(['Product_Category', 'Gender'])['Purchase'].sum().reset_index()

# Pivot for better plotting
pivot_cat_gender = category_gender.pivot(index='Product_Category', columns='Gender', values='Purchase').fillna(0)

# Optional: sort by total
pivot_cat_gender['Total'] = pivot_cat_gender.sum(axis=1)
pivot_cat_gender = pivot_cat_gender.sort_values(by='Total', ascending=False)
pivot_cat_gender.drop(columns='Total', inplace=True)

# Plot
pivot_cat_gender.plot(kind='bar', figsize=(12, 6))
plt.title("Total Purchase by Product Category and Gender")
plt.ylabel("Total Purchase ($)")
plt.xlabel("Product Category")
plt.legend(title='Gender')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Product Category 1 is the highest-grossing category overall, with both men and women spending heavily — but men outspent women by a wide margin.

Across every product category, men consistently outspend women.
There is no category where women outspent men.

The top three categories in terms of spending:

Category 1

Category 5

Category 8
(These likely represent high-volume or high-ticket items — e.g., electronics, appliances, etc.)



```python
## By City Category: 

city_purchase = df.groupby('City_Category')['Purchase'].sum().sort_values(ascending=False)
print(city_purchase)

sns.barplot(x=city_purchase.index, y=city_purchase.values)
plt.title("Total Purchase by City Category")
plt.ylabel("Total Purchase ($)")
plt.xlabel("City Category")
plt.show()
```

```python
city_gender_purchase = df.groupby(['City_Category', 'Gender'])['Purchase'].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=city_gender_purchase, x='City_Category', y='Purchase', hue='Gender')
plt.title("Total Purchase by City Category and Gender")
plt.ylabel("Total Purchase ($)")
plt.xlabel("City Category")
plt.legend(title='Gender')
plt.show()
```

City Category B leads in total purchase volume, followed by Category C, and then Category A.

This suggests Walmart has the highest revenue concentration in City B regions, which could represent Tier 2 or emerging metro cities.

In all city categories, male customers spend significantly more than female customers — a consistent pattern across locations.

The gender gap is largest in City B, which also has the highest overall sales.
City A shows the lowest total purchases for both genders, possibly indicating a smaller customer base or market saturation.

```python
#Loyalty and Tenure: 

#Total Purchase by Stay Duration: 

loyalty_purchase = df.groupby('Stay_In_Current_City_Years')['Purchase'].sum().sort_index()
print(loyalty_purchase)

# Plot
sns.barplot(x=loyalty_purchase.index, y=loyalty_purchase.values)
plt.title("Total Purchase by Years in Current City")
plt.xlabel("Years in Current City")
plt.ylabel("Total Purchase ($)")
plt.show()
```

Customers who have lived in the city for 1 year have the highest total purchase amount.

Beyond 1 year, there’s a drop in total spending as tenure increases (2, 3, and 4+ years).

Surprisingly, newer customers (0–1 years) tend to spend more than those with longer city tenure.

**Insight 1** Male Customers Drive the Majority of Revenue
Men account for significantly more transactions and spending than women — nearly 3x in both volume and value.

Average and median purchase amounts are higher for men as well.

📌 Recommendation:
🎯 Focus Black Friday campaigns and high-ticket product promotions (e.g., electronics, gadgets) toward male customers, who exhibit higher engagement and spending behavior.

**Insight 2** The Age Group 26–35 Is the Biggest Spender
Customers aged 26–35 dominate total purchases, followed by 36–45 and 18–25.

In every age group, men outspent women, reinforcing gender-based targeting.

📌 Recommendation:
👥 Prioritize marketing efforts toward men aged 26–45. Consider segmenting offers based on life stage — e.g., career starters vs. family builders.

**Insight 3:** Product Category Spending Skews Male
Across all product categories, men outspent women — even in categories where women are often assumed to dominate.

Categories 1, 5, and 8 brought in the most revenue overall.

📌 Recommendation:
🛒 Audit the product mix and promotional strategy to better appeal to female shoppers. Consider bundling, targeted recommendations, or exclusive offers in underperforming categories.

**Insight 4:** City Category B Is the Highest-Grossing Market
City Category B had the highest overall spending, followed by C, then A.

Male dominance in spending holds across all city categories.

📌 Recommendation:
🌆 Allocate more inventory and campaign budget to City B, which shows the strongest revenue potential. Explore city-specific offers or pop-up events.

**Insight 5:** Longer Stay in City ≠ Higher Spend
Surprisingly, customers with 1 year in their city spent the most, while 3+ year customers spent less.

Loyalty (as measured by city tenure) does not strongly predict purchase value.

📌 Recommendation:
📦 Consider targeting new movers (0–1 year) with onboarding offers or starter kits. Loyalty strategies should be based on purchase behavior, not just city tenure.
