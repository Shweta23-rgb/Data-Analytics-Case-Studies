### STEP 1

**Business Hypothesis:**
We believe that shortening the sign-up form (shown to Group B) will reduce user friction and lead to more users signing up compared to the current longer form (Group A).

**Primary Metric:**
Sign-Up Rate = Number of users who signed up / Total users in the group

This is a binary metric (yes = 1, no = 0), so we’ll compare proportions using a z-test later on.

**Statistical Hypotheses:**
Null Hypothesis (H₀):
There is no difference in sign-up rates between the two groups.

Conversion_A = Conversion_B

Alternative Hypothesis (H₁):
There is a difference in sign-up rates between the two groups.

Conversion_A ≠ Conversion_B

This will be a two-tailed test, because we’re testing for any difference (not just improvement).



### STEP 2

**Power Analysis:**
Goal: Make sure the test can detect a real difference if it exists.

Power (1 - β): 80% (standard)
This means we accept a 20% chance of missing a real effect (Type II error).

Significance Level (α): 0.05 (5%)
We accept a 5% chance of false positive (Type I error).

**2. Minimum Detectable Effect (MDE)**
This is the smallest difference in sign-up rates that we care about detecting.

Let’s say the business cares about at least a 2% absolute improvement (from 12% to 14%)

So, MDE = 2%

**3. Sample Size (Pre-Defined)**
Based on our assumptions (α = 0.05, Power = 0.8, MDE = 2%), an online calculator gives us approximately:

~2,400 users per group

Total = 4,800 users (we generated 5,000 users in our synthetic dataset — this covers our requirement!)

**4. Test Duration:**
Assume:

Website gets 1,000 users/day
Estimated test duration = 5 days

(We don’t need to simulate time here — just important to keep in mind for real-world experiments.)

### STEP 3

Group A (Control):

Total users: 2,504

Sign-ups: 306

Conversion rate: 12.22%

Group B (Treatment):

Total users: 2,496

Sign-ups: 323

Conversion rate: 12.94%

There’s a slight improvement in Group B's conversion rate — about a 0.72% lift.



```python
# Calculating the signup rate (conversion rate) for each group
conversion_A = df[df['group'] == 'A']['signed_up'].mean() * 100
conversion_B = df[df['group'] == 'B']['signed_up'].mean() * 100

print(f"Signup Rate for Group A: {conversion_A:.2f}%")
print(f"Signup Rate for Group B: {conversion_B:.2f}%")
```

### STEP 4: 

We’ve seen that:

Group A: 12.22% sign-up rate

Group B: 12.94% sign-up rate

That’s a small lift — but is it significant, or could it have happened just by luck?

To find out, we’ll use a z-test for proportions.

```python
from statsmodels.stats.proportion import proportions_ztest

# Getting the counts of signups and total users
signups = [df[df['group'] == 'A']['signed_up'].sum(), df[df['group'] == 'B']['signed_up'].sum()]
totals = [df[df['group'] == 'A']['signed_up'].count(), df[df['group'] == 'B']['signed_up'].count()]

# Running a two-tailed z-test
z_stat, p_value = proportions_ztest(count=signups, nobs=totals, alternative='two-sided')

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
```

Out  p-value here (0.4426) is much greater than the standard threshold of 0.05

This means the difference between Group A and Group B's sign-up rates is not statistically significant

In other words: The 0.72% lift in Group B could very well be due to random chance

**Conclusion:**

We cannot reject the null hypothesis.
There isn’t enough statistical evidence to say that the shorter sign-up form (Group B) performs differently from the original (Group A).
