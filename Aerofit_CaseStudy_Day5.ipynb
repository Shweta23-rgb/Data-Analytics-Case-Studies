```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age
sns.boxplot(x='Product', y='Age', data=df)
plt.title("Age Distribution by Treadmill Type")

# Income
sns.boxplot(x='Product', y='Income', data=df)

# Gender
pd.crosstab(df['Product'], df['Gender'], normalize='index').plot(kind='bar', stacked=True)
```

```python
Table = pd.crosstab(df['Product'], df['MaritalStatus'], normalize='index')
print(Table)


#pd.crosstab() creates a contingency table that summarizes the relationship between two (or more) categorical variables.
#It’s basically like a pivot table in Excel — it counts how often combinations of values occur in your data
```

```python
table_prob = pd.crosstab(df['Product'], df['MaritalStatus'], normalize='index')
table_prob
```
